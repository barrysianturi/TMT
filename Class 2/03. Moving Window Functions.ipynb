```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(600,3),
                  index=pd.date_range('24 Aug 2022',freq='H',periods=600),columns=['A','B','C'])
df.head()
```

```python
df.plot(figsize=(16,9))
plt.show()
```

```python
r = df.rolling(window=10)
r.mean().plot(figsize=(16,9))
```

```python
df.resample('10H').mean()
```

```python
df.resample('10H').mean().plot(figsize=(16,9))
```

```python
df['A'].plot(color='grey',figsize=(16,9),alpha=0.3)
r.mean()['A'].plot(color="red",alpha=0.6)
#df['A'].resample('10H').mean().plot(color='blue',alpha=0.8)
df['A'].ewm(10).mean().plot(color='green',alpha=1)
```

```python
df['A'].expanding(10).mean().plot(figsize=(16,9))
```
